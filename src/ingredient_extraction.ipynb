```python
import numpy as np
import pandas as pd
import os
import json
import sys
```

```python
recipe_path = "../data/recipe_1m_test.json"
with open(recipe_path) as file :
    data = json.load(file)
```

```python
model = FoodModel("chambliss/distilbert-for-food-extraction")
```

```python
exception_file = "../data/food_exceptions.txt"
parsed_dict = {}
for key in data["ingredients"].keys() :
    parsed_list = []
    for texts in data["ingredients"][key] :        
        food_item = model.extract_foods(texts["text"])[0]["Ingredient"]
        if len(food_item) == 0 and ("water" in texts['text']) :
            extracted_food_item = "water"
        elif len(food_item) == 0:
            with open(exception_file, 'a') as file :
                file.write(f'{texts["text"]}\n')
            continue
        else :
            extracted_food_item = model.extract_foods(texts["text"])[0]["Ingredient"][0]["text"]
        parsed_list.append({'text' : extracted_food_item})
    print(f'{key}/{list(data["ingredients"])[-1]}')
    parsed_dict[key] = parsed_list
```

```python
parsed_dict
```

```python
data['extracted ingredients'] = parsed_dict
```

```python
file_name = "../data/recipe_1m_ingredients_extracted.json"
with open(file_name, 'w', encoding='utf-8') as file:
    json.dump(data, file, ensure_ascii=False, indent=4)
```
